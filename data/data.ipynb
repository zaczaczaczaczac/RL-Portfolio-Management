# Define Paths

```python
root_path = "/content/drive/MyDrive/Colab Notebooks/ECE1508_RL/ECE 1508 RL Project/project_files/data"
train_path = os.path.join(root_path, "train")
train_hour_path = os.path.join(train_path, "hourly")
train_day_path = os.path.join(train_path, "daily")
test_path = os.path.join(root_path, "test")
test_hour_path = os.path.join(test_path, "hourly")
test_day_path = os.path.join(test_path, "daily")

ETF_list = ["SPY", "TLT", "GLD", "QQQ"]
```

# Get daily data from

```python
for etf in ETF_list:
  ticker = yf.Ticker(etf)
  train_df = ticker.history(start="2010-01-01", end="2022-12-31", interval="1d")
  train_df = train_df.drop(columns=["Dividends", "Stock Splits", "Capital Gains"], errors="ignore")
  train_df.to_csv(os.path.join(train_day_path, f"{etf}_2010_2022_daily.csv"), index=True)

  test_df = ticker.history(start="2023-01-01", end="2024-12-31", interval="1d")
  test_df = test_df.drop(columns=["Dividends", "Stock Splits", "Capital Gains"], errors="ignore")
  test_df.to_csv(os.path.join(test_day_path, f"{etf}_2023_2024_daily.csv"), index=True)
```

# Get hourly data

```python
for etf in ETF_list:
  ticker = yf.Ticker(etf)
  train_df = ticker.history(start="2024-01-01", end="2024-12-31", interval="1h")
  train_df = train_df.drop(columns=["Dividends", "Stock Splits", "Capital Gains"], errors="ignore")
  train_df.to_csv(os.path.join(train_hour_path, f"{etf}_2024_hourly.csv"), index=True)

  test_df = ticker.history(start="2025-01-01", end="2025-10-31", interval="1h")
  test_df = test_df.drop(columns=["Dividends", "Stock Splits", "Capital Gains"], errors="ignore")
  test_df.to_csv(os.path.join(test_hour_path, f"{etf}_2025_hourly.csv"), index=True)
```
